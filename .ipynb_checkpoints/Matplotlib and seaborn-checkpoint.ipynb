```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#sine wave 

plt.plot(x,y)
plt.show()
```

```python
#cosine wave

plt.plot(x,z)
plt.show()
```

```python
#adding title and x-y axix titles

plt.plot(x,y)
plt.xlabel("Angle x")
plt.ylabel("Angle y")
plt.title("Sine wave graph")
plt.show()
```

```python
#plot a parabola

x = np.linspace(-10,20,40)
y=x**2
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Parabola graph")
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x = np.linspace(-5,10,100)
plt.plot(x,np.sin(x),'g.')
plt.plot(x,np.cos(x),'r.')
plt.show()
```

```python
#Bar Plot
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','hindi','Gujarati','Urdu']
peoples = [100,200,130,450]

ax.bar(languages,peoples)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','hindi','Gujarati','Urdu']
peoples = [100,200,130,450]

ax.pie(peoples , labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
#scatter plot

x=np.linspace(2,10,100)
y=np.sin(x)
z=np.cos(x)

fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='green')
ax.scatter(x,z,color='red')
plt.show()
```

```python
#3d scatter plot

fig3=plt.figure()
ax=plt.axes(projection = '3d')

z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')

plt.show()
```

```python
#Data Visualzation using seaborn

import seaborn as sns 
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_theme()
```

```python
sns.relplot(data=tips , x='total_bill' , y='tip' , col='time' , hue ='smoker' , style='smoker',size='size')
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species', data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species', data=iris)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
sns.barplot(x='sex',y='survived',hue='class' ,data=titanic)
```

```python
sns.displot(titanic['sex'])
```

```python
correlation = titanic.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square = True , fmt='.1f',annot=True , annot_kws = {'size':8} , cmap = 'Blues')
```
